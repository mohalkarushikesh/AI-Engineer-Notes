```python
# Import libraries for data manipulation and plotting
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Generate synthetic dataset using make_blobs
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200,     # Total number of points
                  n_features=2,      # Two features for easy 2D visualization
                  centers=4,         # Create 4 clusters
                  cluster_std=1.8,   # How spread out the clusters are
                  random_state=101)  # Reproducibility
```

```python
# Visualize the synthetic data 
plt.scatter(data[0] [:, 0], data[0] [:, 1], c=data[1], cmap='rainbow')
```

```python
# Apply KMeans Clustering
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
# Fit the model 
kmeans.fit(data[0])
```

```python
# View the coordinates of the discovered cluster centers
kmeans.cluster_centers_
```

```python
# View the predicted cluster labels from KMeans
kmeans.labels_
```

```python
# KMeans results vs. original cluster labels
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:, 1], c = kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:, 1], c = data[1], cmap='rainbow')

plt.show()
```
