```python
# Import necessary libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
df = pd.read_csv('USA_Housing.csv')
```

```python
# Display basic dataset info
df.head() # Show first few rows
```

```python
df.info() # Dataset structure
```

```python
df.describe()  # Summary statistics
```

```python
df.columns  # Column names
```

```python
df['Price'].describe()  # Shows min, max, mean, etc.
```

```python
sns.histplot(x=df['Price'], kde=True, bins=30, color="blue")
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('House Price Distribution')
plt.xlim(df['Price'].min(), df['Price'].max())  # Ensures correct scaling
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')  # Correlation heatmap
```

```python
# Define features (X) and target variable (y)
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
        'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
# Initialize and train the Linear Regression model
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Model parameters
print("Intercept:", lm.intercept_)
print("Coefficients:", lm.coef_)
```

```python
# Create a DataFrame for coefficients
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coeff'])
print(cdf)
```

```python
# Predictions on test set
predictions = lm.predict(X_test)
```

```python
colors = np.abs(y_test - predictions)  # Color based on prediction error
plt.figure(figsize=(8,6))
plt.scatter(y_test, predictions, c=colors, cmap='coolwarm', edgecolors='black', alpha=0.6)
plt.colorbar(label="Prediction Error")
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices with Error Highlight")
plt.grid(True)
plt.show()
```
